```python
# Filter the dataset where trackable_type is "Condition" and get unique trackable_names
conditions = df[df['trackable_type'] == 'Condition']['trackable_name'].unique()
conditions
```

```python
# Define a list of known autoimmune diseases based on the trackable_name column
autoimmune_diseases = [
    'Ulcerative colitis', "Crohn's disease", 'Rheumatoid arthritis',
    'Celiac disease', "Raynaud's disease", 'Sjgorens ',
    'Systemic Scleroderma', 'Type 1 diabetes', "Hashimoto's disease",
    'Psoriasis', 'Myasthenia gravis', 'Undifferentiated Connective Tissue Disease',
    'Polymyositis', 'Vasculitis', 'Sarcoidosis'
]

# Create a new binary column 'autoimmune_disease' where 1 indicates presence of an autoimmune disease
df['autoimmune_disease'] = df['trackable_name'].isin(autoimmune_diseases).astype(int)

# Display the first few rows to verify
df[['trackable_name', 'autoimmune_disease']].head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
```

```python
# Select relevant features
features = ['age', 'sex', 'country']
target = 'autoimmune_disease'
```

```python
# Drop rows with missing target values
df = df.dropna(subset=[target])
```

```python
# Separate features (X) and target (y)
X = df[features]
y = df[target]
```

```python
# Define the preprocessing steps
numeric_features = ['age']
categorical_features = ['sex', 'country']
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
# Define the model pipeline
model = Pipeline(steps=[('preprocessor', preprocessor),
                        ('classifier', LogisticRegression(random_state=42))])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python


# Handle class imbalance using RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X_train, y_train)
```

```python
# Train the model on the resampled data
model.fit(X_resampled, y_resampled)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
```

```python
# Calculate accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)

# Calculate ROC-AUC
roc_auc_rf = roc_auc_score(y_test, y_pred_rf)

# Generate a classification report
classification_rep_rf = classification_report(y_test, y_pred_rf)

# Print the evaluation metrics
print("Accuracy:", accuracy_rf)
print("ROC-AUC:", roc_auc_rf)
print("Classification Report:\n", classification_rep_rf)
```
